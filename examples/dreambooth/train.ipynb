```python
SD_NAME="CompVis/stable-diffusion-v1-4"
TOKEN="aabbccddeeffgg"
INPUT_DIR="/home/ubuntu/data"
MODEL_DIR="/home/ubuntu/model"
PRED_DIR="/home/ubuntu/predict"

!mkdir -p "$MODEL_DIR"
!mkdir -p "$PRED_DIR"

# 1xA100:
BATCH_SIZE=2
LR=1e-6

!(USE_TF32=on python train_dreambooth.py \
    --pretrained_model_name_or_path="$SD_NAME" \
    --instance_data_dir="$INPUT_DIR" \
    --train_text_encoder \
    --output_dir="$MODEL_DIR" \
    --instance_prompt="a photo of ""$TOKEN"" person" \
    --resolution=512 \
    --train_batch_size="$BATCH_SIZE" \
    --gradient_accumulation_steps=1 \
    --lr_scheduler="constant" \
    --lr_warmup_steps=0 \
    --learning_rate="$LR" \
    --max_train_steps=1000 \
    --pred_path "$PRED_DIR" \
 )
```
